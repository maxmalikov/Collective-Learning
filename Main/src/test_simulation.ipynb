```python
list_df = []
for i in range(100):
    model = CollectiveLearningModel(
        n_agents=100,
        n_choices=3,
        seed=i
    )
    for i in range(1):
        model.step()
    df = model.datacollector.get_model_vars_dataframe() 
    list_df.append(df)
df = pd.concat(list_df, ignore_index=True)
```

```python
fig, ax = plt.subplots()
sns.barplot(y=df["Consensus"], ax=ax, errorbar="sd")
ax.set_xlabel("Step")
ax.set_ylabel("Consensus")
ax.set_ylim(0, 1)
ax.grid()
plt.show()
```
